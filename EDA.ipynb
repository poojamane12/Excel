```python
order= df.groupby(['Order ID'], as_index=False)['Amount',].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = order, x = 'Order ID',y= 'Amount',)
```

```python
top 10 order id based on profit
```

```python
order= df.groupby(['Order ID'], as_index=False)['Profit',].sum().sort_values(by='Profit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = order, x = 'Order ID',y= 'Profit',)
```
